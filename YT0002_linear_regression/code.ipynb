```python
import matplotlib.pyplot as plt

def show_plot(x:list, y:list, A:float, B: float):
    fig, ax = plt.subplots()
    
    y_pred = [A * xi + B for xi in x]

    ax.scatter(x, y, color='blue', label='Data')
    ax.plot(x, y_pred, color='red', label='Fitness line')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Linear regression')
    ax.legend()
    return fig, ax
```

```python
def linear_regression(x:list, y:list):
    assert len(x) == len(y), "X and Y have to be the same lenght"

    # Calculate components
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum([x[i] * y[i] for i in range(n)])
    sum_x2 = sum([xi ** 2 for xi in x])

    # Calculate A and B
    A = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
    B = (sum_y - A * sum_x) / n

    return A, B
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

A, B = linear_regression(x, y)

print(f"Values: A = {A}, B = {B}")
```

```python
fig, ax = show_plot(x, y, A, B)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Sample data
x = [[1], [2], [3], [4], [5]]
y = [2, 4, 5, 4, 5]
```

```python
# Create amd fit a linear regression model
model = LinearRegression()
model.fit(x, y)
```

```python
# Getting A and B coefficients
A = model.coef_[0]
B = model.intercept_

print(f"Values: A = {A}, B = {B}")
fig, ax = show_plot([i[0] for i in x], y, A, B)

plt.show()
```

```python
# Prediction for new values ​​of x
x_pred = [[6], [7], [8]]
y_pred = model.predict(x_pred)

fig, ax = show_plot([i[0] for i in x], y, A, B)
ax.scatter([i[0] for i in  x_pred], list(y_pred), color='green', label='New Data')
ax.legend()

plt.show()
```
